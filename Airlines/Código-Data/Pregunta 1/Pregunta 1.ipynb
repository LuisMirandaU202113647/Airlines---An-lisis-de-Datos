```python
#Importar data
import pandas
import pandasql as ps
import seaborn
import numpy as np
import matplotlib.pyplot as plt
cleaned_air=pandas.read_csv('cleaned_airlines.csv')
cleaned_air[:5]
```

```python
#¿Cuáles son los tipos de aviones más populares entre aerolíneas publicadas y aerolíneas operativas?
##Cantidad de tipos de aviones por aerolínea publicada
ps.sqldf('SELECT  "Published Airline", count(DISTINCT "Aircraft Body Type") as "Body Type Count" FROM cleaned_air GROUP BY  "Published Airline" ORDER BY "Body Type Count" DESC',globals())
```

```python
##Cantidad de tipos de aviones por aerolínea operativa
ps.sqldf('SELECT  "Operating Airline", count(DISTINCT "Aircraft Body Type") as "Body Type Count" FROM cleaned_air GROUP BY  "Operating Airline" ORDER BY "Body Type Count" DESC',globals())
```

```python
##Aerolíneas operativaa con 3 tipos de aviones
ps.sqldf('SELECT  "Operating Airline", count(DISTINCT "Aircraft Body Type") as "Body Type Count" FROM cleaned_air GROUP BY  "Operating Airline" HAVING "Body Type Count"==3 ORDER BY "Body Type Count" DESC',globals())
```

```python
##Tipo de avión con mayor cantidad por aerolínea publicada
ps.sqldf('SELECT  "Published Airline", "Aircraft Body Type", Max(Cantidad) as Max_count FROM  ( SELECT  "Published Airline", "Aircraft Body Type", count(*) as Cantidad FROM cleaned_air GROUP BY  "Published Airline", "Aircraft Body Type" ) GROUP BY  "Published Airline" ORDER BY Max_count DESC',globals())
```

```python
##Tipo de avión con mayor cantidad por aerolínea operativa
ps.sqldf('SELECT  "Operating Airline", "Aircraft Body Type", Max(Cantidad) AS Max_count FROM  ( SELECT  "Operating Airline", "Aircraft Body Type", count(*) as Cantidad FROM cleaned_air GROUP BY  "Operating Airline", "Aircraft Body Type" ) GROUP BY  "Operating Airline" ORDER BY Max_count DESC',globals())
```

```python
##Tipos de avión y la cantidad de aerolíneas publicads que los tienen como más populares
aircraft_body_published=ps.sqldf('SELECT  "Aircraft Body Type", count("Published Airline") as "Published Airline Count" FROM ( SELECT  "Published Airline", "Aircraft Body Type", Max(Cantidad) FROM  ( SELECT  "Published Airline", "Aircraft Body Type", count(*) as Cantidad FROM cleaned_air GROUP BY  "Published Airline", "Aircraft Body Type" ) GROUP BY  "Published Airline" ) GROUP BY  "Aircraft Body Type" ORDER BY  "Published Airline Count" DESC',globals())
aircraft_body_published
```

```python
##Exportación del dataframe
aircraft_body_published.to_csv('aircraft_body_published.csv')
```

```python
##Tipos de avión y la cantidad de aerolíneas operativas que los tienen como más populares
aircraft_body_operating=ps.sqldf('SELECT  "Aircraft Body Type", count("Operating Airline") as "Operating Airline Count" FROM ( SELECT  "Operating Airline", "Aircraft Body Type", Max(Cantidad) FROM  ( SELECT  "Operating Airline", "Aircraft Body Type", count(*) as Cantidad FROM cleaned_air GROUP BY  "Operating Airline", "Aircraft Body Type" ) GROUP BY  "Operating Airline" ) GROUP BY  "Aircraft Body Type" ORDER BY  "Operating Airline Count" DESC',globals())
aircraft_body_operating
```

```python
##Exportación del dataframe
aircraft_body_operating.to_csv('aircraft_body_operating.csv')
```
