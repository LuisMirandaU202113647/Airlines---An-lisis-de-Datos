```python
import pandas
import pandasql as ps
import seaborn
import matplotlib 
cleaned_air=pandas.read_csv('cleaned_airlines.csv')
cleaned_air[:5]
```

```python
#Exploración de datos relevantes
##Obtener aerolíneas operativas, aerolíneas publicadas y peso total desembarcado por manufacturero
manufacturer_airlines=ps.sqldf('SELECT DISTINCT  "Aircraft Manufacturer" AS Aircraft_Manufacturer FROM  cleaned_air',globals())
manufacturer_operating_airlines=ps.sqldf('SELECT DISTINCT  "Aircraft Manufacturer" AS Aircraft_Manufacturer, "Operating Airline" AS Operating_Airline FROM  cleaned_air ',globals())
manufacturer_published_airlines=ps.sqldf('SELECT  DISTINCT "Aircraft Manufacturer" AS Aircraft_Manufacturer, "Published Airline" AS Published_Airline FROM  cleaned_air ',globals())
manufacturer_airlines
```

```python
##Cantidad de aerolíneas operativas por manufacturero
ps.sqldf('  SELECT Aircraft_Manufacturer, count(*) as "Operating Airline Count" FROM manufacturer_operating_airlines GROUP BY Aircraft_Manufacturer ORDER BY "Operating Airline Count" DESC',globals())
```

```python
##Cantidad de aerolineas publicadas por manufacturero
ps.sqldf('SELECT Aircraft_Manufacturer, count(*) as "Published Airline Count" FROM manufacturer_published_airlines GROUP BY Aircraft_Manufacturer ORDER BY "Published Airline Count" DESC',globals())
```

```python
##Cantidad de solo aerolineas operativas por manufacturero
only_operating_airline=ps.sqldf('SELECT oa.Aircraft_Manufacturer, count(*) AS Only_Operating_Airline_Count FROM manufacturer_operating_airlines oa LEFT JOIN manufacturer_published_airlines pa ON  oa.Aircraft_Manufacturer=pa.Aircraft_Manufacturer AND oa.Operating_Airline=pa.Published_Airline WHERE Published_Airline IS NULL GROUP BY oa.Aircraft_Manufacturer',globals())
only_operating_airline
```

```python
##Cantidad de solo aerolineas publicadas por manufacturero
only_published_airline=ps.sqldf('SELECT pa.Aircraft_Manufacturer, count(*) as Only_Published_Airline_Count FROM manufacturer_published_airlines pa LEFT JOIN manufacturer_operating_airlines oa ON  pa.Aircraft_Manufacturer=oa.Aircraft_Manufacturer AND pa.Published_Airline=oa.Operating_Airline WHERE oa.Operating_Airline IS NULL GROUP BY pa.Aircraft_Manufacturer',globals())
only_published_airline
```

```python
##Cantidad de aerolineas que son operativas y publicadas por manufacturero
operating_published_airline=ps.sqldf('SELECT pa.Aircraft_Manufacturer, count(*) as Published_and_Operating_Airline_Count FROM manufacturer_published_airlines pa INNER JOIN manufacturer_operating_airlines oa ON  pa.Aircraft_Manufacturer=oa.Aircraft_Manufacturer AND pa.Published_Airline=oa.Operating_Airline GROUP BY pa.Aircraft_Manufacturer ORDER BY "Published and Operating Airline Count" DESC',globals())
operating_published_airline
```

```python
##Peso total por manufacturero
total_weight=ps.sqldf(' SELECT  "Aircraft Manufacturer" AS Aircraft_Manufacturer, sum("Total Landed Weight") AS Total_Landed_Weight FROM cleaned_air GROUP BY  Aircraft_Manufacturer ORDER BY Total_Landed_Weight DESC',globals())
total_weight
```

```python
#Estadisticas completas
est_completas=ps.sqldf(' SELECT  ma.Aircraft_Manufacturer, oa.Only_Operating_Airline_Count, pa.Only_Published_Airline_Count, opa.Published_and_Operating_Airline_Count, tw.Total_Landed_Weight FROM manufacturer_airlines ma LEFT JOIN  only_operating_airline oa ON  ma.Aircraft_Manufacturer=oa.Aircraft_Manufacturer LEFT JOIN  only_published_airline pa ON ma.Aircraft_Manufacturer=pa.Aircraft_Manufacturer LEFT JOIN  operating_published_airline opa ON ma.Aircraft_Manufacturer=opa.Aircraft_Manufacturer INNER JOIN total_weight tw ON ma.Aircraft_Manufacturer=tw.Aircraft_Manufacturer',globals())
est_completas
```

```python
###Reemplazar valores nulos
est_completas.fillna(0,inplace=True)
est_completas
```

```python
##Obtener las estadísticas finales
est_finales=ps.sqldf(' SELECT  Aircraft_Manufacturer, Only_Operating_Airline_Count, Only_Published_Airline_Count, Published_and_Operating_Airline_Count, Only_Operating_Airline_Count+Only_Published_Airline_Count+Published_and_Operating_Airline_Count AS Total_airlines_count, Total_Landed_Weight, Total_Landed_Weight/(Only_Operating_Airline_Count+Only_Published_Airline_Count+Published_and_Operating_Airline_Count) AS Average_Landed_Weight FROM est_completas ORDER BY Total_airlines_count DESC',globals())
est_finales
```

```python
##Importar estadísticas a csv
est_finales.to_csv('est_finales.csv',sep=';',encoding='utf-8')
```
